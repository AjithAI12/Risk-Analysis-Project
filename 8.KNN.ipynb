```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as matlab
```

```python
dataset=pd.read_csv("preprocessed_data.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
dataset["LeaveOrNot"].value_counts()
```

```python
independent=dataset[['JoiningYear', 'PaymentTier', 'Age', 'ExperienceInCurrentDomain',
        'Education_Masters', 'Education_PHD', 'City_New Delhi',
       'City_Pune', 'Gender_Male', 'EverBenched_Yes']]
```

```python
independent
```

```python
dependent=dataset[["LeaveOrNot"]]
dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': [3, 5, 7, 9],'weights': ['uniform', 'distance'],'p': [1, 2] } 



grid = GridSearchCV(KNeighborsClassifier(), param_grid, refit = True, verbose = 3,n_jobs=1,scoring='f1_weighted') 
grid.fit(X_train, Y_train) 
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(X_test) 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, grid_predictions)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, grid_predictions)
```

```python
print(grid.best_params_)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
